```python
# Replace missing Income values with the median income
median_income = df["Income"].median()
df["Income"] = df["Income"].fillna(median_income)

# Verify that missing values are gone
df["Income"].isnull().sum()
```

```python
# ---- 1. AGE ----
df["Age"] = 2024 - df["Year_Birth"]

# ---- 2. TOTAL SPEND ----
spend_cols = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts",
              "MntSweetProducts","MntGoldProds"]

df["Total_Spend"] = df[spend_cols].sum(axis=1)

# ---- 3. FAMILY SIZE ----
df["Family_Size"] = df["Kidhome"] + df["Teenhome"] + 2   # +2 for parents

# ---- 4. CUSTOMER SINCE ----
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
df["Customer_For_Days"] = (pd.to_datetime("today") - df["Dt_Customer"]).dt.days

# ---- 5. TOTAL PURCHASES ----
purchase_cols = ["NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases"]
df["Total_Purchases"] = df[purchase_cols].sum(axis=1)

# Preview new columns
df[["Age","Total_Spend","Family_Size","Customer_For_Days","Total_Purchases"]].head()
```

```python
# ---- 1. AGE (create only if not exists) ----
if "Age" not in df.columns:
    df["Age"] = 2024 - df["Year_Birth"]

# ---- 2. TOTAL SPEND ----
spend_cols = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts",
              "MntSweetProducts","MntGoldProds"]

if "Total_Spend" not in df.columns:
    df["Total_Spend"] = df[spend_cols].sum(axis=1)

# ---- 3. FAMILY SIZE ----
if "Family_Size" not in df.columns:
    df["Family_Size"] = df["Kidhome"] + df["Teenhome"] + 2   # +2 for parents

# ---- 4. CUSTOMER SINCE ----
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
df["Customer_For_Days"] = (pd.to_datetime("today") - df["Dt_Customer"]).dt.days

# ---- 5. TOTAL PURCHASES ----
purchase_cols = ["NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases"]
df["Total_Purchases"] = df[purchase_cols].sum(axis=1)

# Preview new columns
df[["Age","Total_Spend","Family_Size","Customer_For_Days","Total_Purchases"]].head()
```

```python
# Fix for Dt_Customer parsing and compute Customer_For_Days safely

# Try parsing with dayfirst=True
df['Dt_Customer_parsed'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True, errors='coerce')

# Count bad parses
bad_parse_count = df['Dt_Customer_parsed'].isna().sum()
print(f"Rows failing to parse with dayfirst=True: {bad_parse_count}")

# If some failed, try a fallback parse with common alternative (YYYY-MM-DD)
if bad_parse_count > 0:
    mask_bad = df['Dt_Customer_parsed'].isna()
    df.loc[mask_bad, 'Dt_Customer_parsed'] = pd.to_datetime(
        df.loc[mask_bad, 'Dt_Customer'], errors='coerce'
    )
    bad_parse_count2 = df['Dt_Customer_parsed'].isna().sum()
    print(f"Rows still failing after fallback: {bad_parse_count2}")

# Now compute Customer_For_Days (days since Dt_Customer_parsed)
today = pd.to_datetime('today')
df['Customer_For_Days'] = (today - df['Dt_Customer_parsed']).dt.days

# If any Customer_For_Days are NaN (because parse failed), fill them with the median
nan_days = df['Customer_For_Days'].isna().sum()
print(f"Rows with NaN Customer_For_Days before filling: {nan_days}")

if nan_days > 0:
    median_days = int(df['Customer_For_Days'].median(skipna=True))
    df['Customer_For_Days'] = df['Customer_For_Days'].fillna(median_days)
    print(f"Filled NaN Customer_For_Days with median = {median_days}")

# Replace Dt_Customer with the parsed column (optional)
df['Dt_Customer'] = df['Dt_Customer_parsed']
df = df.drop(columns=['Dt_Customer_parsed'])

# Quick preview of the column
print(df[['Dt_Customer','Customer_For_Days']].head())
print("\nDone parsing and computing Customer_For_Days.")
```

```python

import pandas as pd

# Load the scaled features from Week 1
X_scaled = pd.read_csv("X_scaled_week1.csv")

# Load cleaned dataset
df_clean = pd.read_csv("customers_clean_week1.csv")

X_scaled.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd

inertia_values = []
silhouette_values = []
k_values = range(2, 11)  # Testing k = 2 to 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)

    inertia_values.append(kmeans.inertia_)
    silhouette_values.append(silhouette_score(X_scaled, labels))

# Create a results table
results = pd.DataFrame({
    "k": k_values,
    "Inertia": inertia_values,
    "Silhouette": silhouette_values
})

results
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(results["k"], results["Inertia"], marker='o')
plt.title("Elbow Method - Inertia vs k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(results["k"], results["Silhouette"], marker='o', color='orange')
plt.title("Silhouette Score vs k")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()
```

```python
# Find k with highest silhouette score
best_k = results.loc[results["Silhouette"].idxmax(), "k"]
best_k
```

```python
from sklearn.cluster import KMeans

# Train K-Means with k=2
final_kmeans = KMeans(n_clusters=2, random_state=42)
cluster_labels = final_kmeans.fit_predict(X_scaled)

# Add cluster labels to cleaned dataset
df_clean['Cluster'] = cluster_labels

# Preview the updated dataset
df_clean[['ID', 'Cluster']].head()
```

```python
df_clean.to_csv("customers_with_clusters.csv", index=False)
print("Clustered dataset saved successfully!")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Reduce scaled data to 2 dimensions
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scaled)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(pca_components[:,0], pca_components[:,1], c=cluster_labels, cmap='coolwarm')
plt.title("Customer Segments (PCA Visualization)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(label="Cluster")
plt.show()
```
